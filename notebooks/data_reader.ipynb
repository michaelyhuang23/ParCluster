```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import os
```

```python
root = '../results'
names = ['simden']
```

```python
dict_data = []
data_names = os.listdir(root)
for name in names:
    for data_name in data_names:
        if not name in data_name: continue
        run = int(data_name[-1:])
        print(f'reading {data_name}')
        for item in os.listdir(os.path.join(root, data_name)):
            if not item.endswith('.txt'): continue
            with open(os.path.join(root, data_name, item)) as f:
                lines = f.readlines()
            if len(lines) != 3: continue
            nums = [float(line[:-1]) for line in lines]
            file_name = item.split('.')[0]
            alg_name = file_name.split('_')[0]
            r = int(file_name.split('_')[1])
            n = int(file_name.split('_')[2])
            thread = int(file_name.split('_')[3])
            dict_data.append({'name':alg_name, 'r':r, 'n':n, 'thread':thread, 'run':run, 'part':'density computation', 'time':nums[0]})
            dict_data.append({'name':alg_name, 'r':r, 'n':n, 'thread':thread, 'run':run, 'part':'dependent point finding', 'time':nums[1]})
            dict_data.append({'name':alg_name, 'r':r, 'n':n, 'thread':thread, 'run':run, 'part':'single linkage clustering', 'time':nums[2]})
df = pd.DataFrame.from_records(dict_data)
    
```

```python
df
```

```python
def drop_and_sum(df, column):
    if column not in df.columns: return df
    df = df.drop([column], axis=1)
    columns = list(df.columns)
    columns.remove('time')
    df = df.groupby(columns).sum()
    return df.reset_index()
```

```python
df = drop_and_sum(df, 'run')
df['time']/=3
```

```python
df=df.loc[df['r']==30]
```
